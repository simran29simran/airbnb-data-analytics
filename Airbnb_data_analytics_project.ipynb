```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("compressed_data.csv")
```

# Handling missing values

```python
df['last review'] = pd.to_datetime(df['last review'] , errors = 'coerce')
```

```python
df.info()
```

```python
df.fillna({'reviews per month' : 0 , 'last review': df["last review"].min()}, inplace = True)
```

```python
df.dropna(subset=["NAME" , "host name"] , inplace  = True)
```

```python
print(df.isnull() . sum())
```

```python
df = df.drop(columns=["license", "house_rules"], errors='ignore')
```

```python
df.head()
```

```python
#remove dollar sign and convert to float
df['price'] = df['price'].replace(r'[\$,]', '', regex=True).astype(float)
df['service fee'] = df['service fee'].replace(r'[\$,]', '', regex=True).astype(float)
```

```python
df.head()
```

# Remove duplicates

```python
df.drop_duplicates(inplace = True)
```

```python
df.info()
```

# descriptive statistics

```python
df.describe()
```

# Visualization

# What is the distribution of the listing price

```python
plt.figure(figsize = (10,6))
sns.histplot(df['price'] , bins = 50 , kde = True , color = 'blue')
plt.title("Distribution of the listing price")
plt.xlabel("Price $")
plt.ylabel("Friquency")
```

The Histogram shows a fairly even distribution of listing price across different price ranges ,
indicating no particular concentration of listing  things

# How are diffrent types of rooms ditribute?

```python
df['room type']
```

```python
plt.figure(figsize = (8,5))
sns.countplot(x = 'room type' , data = df  , color = 'hotpink')
plt.title('Room type distribution')
plt.xlabel("Room type")
plt.ylabel("count")
plt.show()
```

# How are listing distributed across different neighbourhood group

```python
plt.figure(figsize = (12 ,8))
sns.countplot( y = "neighbourhood group" , data = df , color = "lightgreen" , order = df['neighbourhood group'].value_counts().index)
plt.title("Number of lighting by Neighbourhood Group")
plt.xlabel("Count")
plt.ylabel("Neighbourhood Group")
plt.show()
```

# What is the relationship between price and room type

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='room type', y='price', hue='room type', data=df, palette='Set1', legend=False)
plt.title("Price vs Room Type")
plt.xlabel("Room Type")
plt.ylabel("Price")
plt.show()
```

# How has the number of review change over the time

```python
df['last review'] = pd.to_datetime(df['last review'])
review_over_time = df.groupby(df['last review'].dt.to_period('M')).size()

plt.figure(figsize=(12,6))
review_over_time.plot(kind='line', color='red')
plt.title("Number of review over the time")
plt.xlabel("Date")
plt.ylabel("Number of review")
plt.show()
```
